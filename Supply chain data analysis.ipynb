```python
#importing necessary library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

EDA Analysis of data to understand the data set we have. ‚õè

```python
#Load data CSV file into the pandas data frame
data= pd.read_csv("/supply_chain_data.csv")
data
```

```python
#clean the data if necessary, first checking for data type of each column of the data ser=t,
data.info()
```

```python
# OR Code shows number of null value/ here in the data we have zero nulls
data.isnull().sum()
```

```python
#Code check for the duplicates/ reult zero
data.duplicated().sum()
```

```python
#Code for data information
data.describe()
```

```python
#histogram for price
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=20, kde=True)
plt.title('Histogram of Price')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
#Univarient Analysis "To uderstand the distrbution of numerical values, checking for outliers"
# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product type', y='Price', data=data)
plt.title('Box Plot of Prices by Product type')
plt.xlabel('Product type')
plt.ylabel('Price')
plt.show()
```

```python
# Now create the bar plot
plt.figure(figsize=(8, 3))
sns.barplot(x='Product type', y='Availability', data=data)
plt.title('Availability by Product type')
plt.xlabel('Product type')
plt.ylabel('Availability')
plt.show()
```

```python
# Now create the bar plot
plt.figure(figsize=(8, 3))
sns.barplot(x='Product type', y='Number of products sold', data=data)
plt.title('Number of Product sold by Product type')
plt.xlabel('Product type')
plt.ylabel('Counts')
plt.show()
```

```python
# Melt the DataFrame to long-form for seaborn
melted_data = data.melt(id_vars='Product type', value_vars=['Availability', 'Number of products sold'],
                        var_name='Metric', value_name='Value')

# Set up the plotting area
plt.figure(figsize=(8, 4))

# Create the bar plot
sns.barplot(x='Product type', y='Value', hue='Metric', data=melted_data)

# Add labels and title
plt.title('Availability and Number of Products Sold by Product Type')
plt.xlabel('Product Type')
plt.ylabel('Counts')
plt.legend(title='Metric')

# Show the plot
plt.show()
```

```python
#Revenu generated based on catogery'
plt.figure(figsize=(10, 6))
sns.boxplot(x='Product type', y='Revenue generated', data=data)
plt.title('Box Plot of Prices by Product type')
plt.xlabel('Product type')
plt.ylabel('Revenue generated')
plt.show()
```

```python
# Now create the bar plot
plt.figure(figsize=(8, 3))
sns.barplot(x='Product type', y='Revenue generated', data=data)
plt.title('Revenue generated by Product type')
plt.xlabel('Product type')
plt.ylabel('Revenue generated')
plt.show()
```
